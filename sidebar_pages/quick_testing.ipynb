```python
def max_square_matrix(matrix):
    rows = len(matrix)
    cols = len(matrix[0])
    matrix_copy = matrix
    max_m = 0

    if 1 in matrix[0]:
        max_m = 1

    for i in range(rows):
        if matrix[i][0] == 1:
            max_m = 1
    
    for i in range(1,rows):
        for j in range(1,cols):
            if matrix[i][j] != 0:
                moment_m = min(matrix_copy[i-1][j],
                matrix_copy[i-1][j-1],matrix_copy[i][j-1])+1
                matrix_copy[i][j] = moment_m
                if moment_m > max_m:
                    max_m = moment_m
    
    return max_m, matrix_copy
```

```python
def permute(N, index, result):
    nums = [0]*N

    if index == len(nums):
        #printAll(nums)
        #print(nums)
        #print(result)
        return result.append(0)

    for i in range(5):
        nums[index] = i
        permute(N,index + 1,result)
    
    return result
    
```

```python
def sum_nums(arr, index):
    if index <= 0:
        return arr
    else:
        arr.append(index)
        return sum_nums(arr, index - 1)
```
