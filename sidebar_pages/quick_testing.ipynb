```python
def max_square_matrix(matrix):
    rows = len(matrix)
    cols = len(matrix[0])
    matrix_copy = matrix
    max_m = 0

    if 1 in matrix[0]:
        max_m = 1

    for i in range(rows):
        if matrix[i][0] == 1:
            max_m = 1
    
    for i in range(1,rows):
        for j in range(1,cols):
            if matrix[i][j] != 0:
                moment_m = min(matrix_copy[i-1][j],
                matrix_copy[i-1][j-1],matrix_copy[i][j-1])+1
                matrix_copy[i][j] = moment_m
                if moment_m > max_m:
                    max_m = moment_m
    
    return max_m, matrix_copy
```

```python
def check0s(arr):
    for i, element in enumerate(arr):
        try:
            if element == 0 and arr[i+1] == 0:
                return True
        except:
            pass
    return False
```

```python
def permute(nums, index, result):
    if index == len(nums):
        #print(nums)
        check_0 = check0s(nums)
        #print(check_0)
        if check_0 == False:
            return result.append(0)
        else:
            return result

    for i in range(5):
        nums[index] = i
        permute(nums,index + 1,result)

    return result
```

```python
def resutl_perms(N):
    nums = [0]*N
    result = permute(nums, 0, [])
    return len(result)
```

```python
def sum_nums(arr, index):
    if index <= 0:
        return arr
    else:
        arr.append(index)
        return sum_nums(arr, index - 1)
```
